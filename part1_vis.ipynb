```python
from part1 import main as part1
from collections import Counter
from matplotlib import pyplot as plt
from nltk import word_tokenize
```

```python
fanfic_rslts, spell_rslts, onion_rslts = part1()
len(fanfic_rslts)
```

```python
fic_tokens = fanfic_rslts[0]

def plot_token_freqs(tokens, title, dps=30):
    x, y = zip(*sorted(Counter(tokens).items(), key=lambda i: i[1], reverse=True)[:dps])
    plt.bar(x, y)
    plt.xticks(rotation=90)
    plt.title(title)

plot_token_freqs(fic_tokens, "Most frequent tokens in fanfiction dataset")
```

```python
fic_stems = fanfic_rslts[1]
plot_token_freqs(fic_stems, "Most frequent stemmed tokens in fanfiction dataset")
```

```python
sents = fanfic_rslts[3]

def plot_sent_lens(sents, title="Sentence length distribution"):
    sentlens = [len(word_tokenize(sent)) for sent in sents]

    plt.hist(sentlens, bins=100)
    plt.title(title)
    plt.xlabel("Sentence Lengths (tokens)")
    plt.ylabel("Frequency (count)")
    plt.ylabel("")

print(sorted(sents, key=len, reverse=True)[0])
plot_sent_lens(sents, "Sentence Length Distribution (fanfiction dataset)")
```
